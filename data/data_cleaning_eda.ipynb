### Handling with Null Values

#### Purchase Amount (USD)

```python
# Fill the null value to the mean of their Purechase Amount (USD) according to Item Purchased
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```

#### Size

```python
df['Size'].unique()
```

```python
df['Size'].value_counts()
```

```python
df.groupby(by=['Category','Item Purchased'])['Size'].unique()
```

```python
# Replacing all values and null value of 'Electronics' Category with 'Not Applicable'
df.loc[df['Category']=='Electronics','Size'] = 'Not Applicable'
```

```python
df.groupby(by='Category')['Size'].unique()
```

```python
# Replacing all value and null values of 'Accessories' Category with 'Free Size'
df.loc[df['Category']=='Accessories','Size'] = 'Free Size'
```

```python
df.groupby(by='Category')['Size'].unique()
```

```python
# Replace all value and null values of 'Footwear' Category with 'Not Available'
df.loc[df['Category']=='Footwear','Size'] = 'Not Available'
```

```python
df.groupby(by='Category')['Size'].unique()
```

```python
# Replacing null values of 'Clothong' Category with Mode values
df.loc[(df['Category']=='Clothing') & (df['Size'].isnull()),'Size'] = df[df['Category']=='Clothing']['Size'].mode()[0]
```

```python
df.groupby(by='Category')['Size'].unique()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'].unique()
```

```python
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
# Fill null values of 'Review Rating' with the mean of each 'Item Purchased'
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

#### Previous Purchases

```python
# Fill the Null value to 0 of 'Previous Purchases'
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```
